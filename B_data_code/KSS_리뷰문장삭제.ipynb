```python
ltk_without_pos = []

for i in tqdm(range(len(ltk2))):
  for token,pos in ltk2[i]:
      if pos in ['Noun', 'Adjective'] and len(token) > 1:
        if token in stopwords3:
          ltk2[i] = ''
        # elif not token in stopwords:
        #   ltk_without_neg.append(token)
        else:
          ltk_without_pos.append(token)
```
